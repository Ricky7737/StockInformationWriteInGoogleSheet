```python
import pandas as pd
```

```python
#外資買賣超資料 : 來自證券交易所
TW_ForInv50_url = 'https://www.twse.com.tw/rwd/zh/fund/TWT38U?response=html'
TW_ForInv50_df = pd.read_html(TW_ForInv50_url)[0]
```

```python
TW_ForInv50_df
```

```python
#擷取需要的資料，去除不需要的欄位與留下前50筆資料
new_TW_ForInv50_df = TW_ForInv50_df.iloc[:50, [1,2,11]]
new_TW_ForInv50_df
```

```python
#重新設定欄位名稱
new_TW_ForInv50_df.columns = ['證券代號','證券名稱','買超股數']
new_TW_ForInv50_df
```

```python
#因為最後一欄資料單位為股數，將資料除1000後取整數換算成張數
# // 整除後取整數
new_TW_ForInv50_df['買超張數'] = new_TW_ForInv50_df['買超股數']//1000
#去除買超股數
new_TW_ForInv50_df
```

```python
new_TW_ForInv50_df = new_TW_ForInv50_df.drop(['買超股數'], axis=1)
```
