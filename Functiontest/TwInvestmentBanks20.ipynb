```python
import pandas as pd
```

```python
# 上市投信買賣超連結
Tw_IvenBank20_url = 'https://www.twse.com.tw/rwd/zh/fund/TWT44U?response=html'
Tw_IvenBank20_df = pd.read_html(Tw_IvenBank20_url)[0]
Tw_IvenBank20_df
```

```python
Tw_IvenBank20_df = Tw_IvenBank20_df.iloc[:20, [1,2,5]]
Tw_IvenBank20_df
```

```python
Tw_IvenBank20_df.columns = ['證券代號','證券名稱','買超股數']
Tw_IvenBank20_df['買超張數'] = Tw_IvenBank20_df['買超股數']//1000
Tw_IvenBank20_df
```

```python
#axis = 1 為指定成欄位
Tw_IvenBank20_df = Tw_IvenBank20_df.drop(['買超股數'], axis=1)
Tw_IvenBank20_df
```
