# 1.載入爬蟲所需要的套件

```python
from bs4 import BeautifulSoup
import requests
import lxml
import pandas as pd
```

```python
# 抓取資料來源，以證交所為例
response = requests.get(
    "https://www.twse.com.tw/zh/trading/foreign/bfi82u.html")
# 引用 beautifulsoup類別，傳入HTML字串結構，並且指定為 HTML 解析型態
soup = BeautifulSoup(response.text, "lxml")
```

```python
# 輸出解析後的網頁 HTMP 格式文件
#prettify() 將soup這個物件美化
print(soup.prettify())
```
